```python
import pandas as pd
```

```python
incident = pd.read_csv("CHDS_raw_Data.xlsx - INCIDENT.csv")
```

```python
incident
```

```python
victim = pd.read_csv("CHDS_raw_Data.xlsx - VICTIM.csv")
```

```python
victim['injury'].unique()
```

```python
fatal_victim = victim[victim['injury'] == "Fatal"]
```

```python
fatal_victim
```

```python
fatal_victim_incident = fatal_victim.merge(incident, how="left",left_on="incidentid", right_on="Incident_ID")
```

```python
fatal_victim_incident
```

```python
fatal_victim_count_temp = fatal_victim_incident.groupby("incidentid").count()
```

```python
fatal_victim_count = fatal_victim_count_temp.reset_index()[['incidentid', 'injury']]
```

```python
fatal_incident = fatal_victim_count.merge(incident, how="left", left_on="incidentid", right_on="Incident_ID")
```

```python
fatal_incident.sort_values(by="injury")
```

```python
nc_mu_fatal_incident = fatal_incident[(fatal_incident['Date'] >= '2000-01-01') & (fatal_incident['injury'] >= 2)]
```

```python
nc_mu_fatal_incident
```

```python
nc_mu_fatal_incident.injury.sum()
```

```python
nc_mu_fatal_incident.columns
```

```python
nc_mu_fatal_incident['Summary'].apply(lambda x: len(x))
```

```python
nc_mu_fatal_incident['Narrative'].dropna().apply(lambda x: len(x))
```

```python
nc_mu_fatal_incident['Narrative'].iloc[1]
```

```python
nc_mu_fatal_incident['date_summary'] = nc_mu_fatal_incident.apply(lambda df: df['Date'] + '\n' + df['Summary'], axis=1)
```

```python
nc_mu_fatal_incident['date_summary']
```

```python
output = nc_mu_fatal_incident[['injury', 'date_summary']]
```

```python
output
```

```python
output.to_csv('data.csv', index=False)
```
